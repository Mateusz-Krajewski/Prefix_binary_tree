```python
import matplotlib.pyplot as plt
import numpy as np

# Dane benchmarków
benchmark1 = {
    "AddElement/100": 2091,
    "AddElement/1000": 117849,
    "AddElement/10000": 10179407,
    "AddElement/100000": 0,
    "CheckFunction/100": 1732,
    "CheckFunction/1000": 113862,
    "CheckFunction/10000": 10438629,
    "CheckFunction/100000": 0,
}

benchmark2 = {
    "AddElement/100": 2593,
    "AddElement/1000": 166491,
    "AddElement/10000": 15573133,
    "AddElement/100000": 1920578766,
    "CheckFunction/100": 2220,
    "CheckFunction/1000": 174185,
    "CheckFunction/10000": 16267424,
    "CheckFunction/100000": 5377693057,
}

benchmark3 = {
    "AddElement/10": 27.7,
    "AddElement/100": 185,
    "AddElement/1000": 1899,
    "AddElement/10000": 17801,
    "AddElement/100000": 166833,
    "CheckFunction/10": 28.4,
    "CheckFunction/100": 194,
    "CheckFunction/1000": 1791,
    "CheckFunction/10000": 17711,
    "CheckFunction/100000": 170009,
}

# Przygotowanie danych do wykresu
labels = list(benchmark1.keys())
time1 = [benchmark1.get(key, 0) for key in labels]
time2 = [benchmark2.get(key, 0) for key in labels]
time3 = [benchmark3.get(key, 0) for key in labels]

x = np.arange(len(labels))  # Pozycje na osi x
width = 0.25  # Szerokość słupków

# Tworzenie wykresu
fig, ax = plt.subplots(figsize=(16, 9))
rects1 = ax.bar(x - width, time1, width, label='static_list', color='steelblue')
rects2 = ax.bar(x, time2, width, label='sorted_list', color='darkorange')
rects3 = ax.bar(x + width, time3, width, label='binary_tree', color='seagreen')

# Dostosowanie wykresu
ax.set_xlabel('Benchmark')
ax.set_ylabel('Time (ns)')
ax.set_title('Porównanie wyników benchmarków')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45, ha="right")
ax.legend()

# Dodawanie wartości nad słupkami
def add_labels(rects, offset=5):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.0f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, offset),
                    textcoords="offset points",
                    fontsize=10,
                    ha='center', va='bottom')

add_labels(rects1, offset=10)
add_labels(rects2, offset=15)
add_labels(rects3, offset=20)

# Skalowanie osi y logarytmicznie (opcjonalnie, jeśli różnice są bardzo duże)
ax.set_yscale('log')

fig.tight_layout()
plt.show()
```
